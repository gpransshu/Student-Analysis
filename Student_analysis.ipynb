```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Read the CSV file into a dataframe
df = pd.read_csv('Student.csv')
```

```python
df.head()
```

```python
# Remove the "Timestamp" column
df = df.drop('Timestamp', axis=1)

# Change the column names to simpler terms
df.columns = ['Gender', 'Age', 'Course', 'Year of Study', 'CGPA', 'Marital Status',
              'Depression', 'Anxiety', 'Panic Attack', 'Seek Treatment']
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True) #Handling Missing Values
```

### Encoding

```python
# Perform one-hot encoding
one_hot_cols = ['Course','Gender', 'Marital Status', 'Depression', 'Anxiety', 'Panic Attack', 'Seek Treatment']
df_encoded = pd.get_dummies(df, columns=one_hot_cols)

# Perform label encoding
label_encode_cols = ['Year of Study']
for col in label_encode_cols:
    df_encoded[col] = df_encoded[col].astype('category').cat.codes

# Numerical encoding
# If there are specific ranges in the 'CGPA' column that hold meaning, you can assign corresponding numerical values

# Print the encoded DataFrame
print(df_encoded)
```

```python

# Calculate the average CGPA and add it as a new column
df['avg_cgpa'] = (df['CGPA'].str.split(' - ')
                      .apply(lambda x: (float(x[0]) + float(x[1])) / 2))
df
```

### Data Visualization

```python
# Histogram - Age
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter Plot - Age vs. CGPA
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['CGPA'])
plt.title('Age vs. CGPA')
plt.xlabel('Age')
plt.ylabel('CGPA')
plt.show()
```

```python
# Pie Chart - Course
course_counts = df['Course'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(course_counts, labels=course_counts.index, autopct='%1.1f%%')
plt.title('Course Distribution')
plt.show()
```

```python
# Heatmap - Correlation
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Box Plot - CGPA
plt.figure(figsize=(10, 6))
sns.boxplot(df['avg_cgpa'])
plt.title('CGPA Distribution')
plt.xlabel('CGPA')
plt.show()
```

```python

# Count the occurrences of each category
depression_counts = df['Depression'].value_counts()
anxiety_counts = df['Anxiety'].value_counts()
panic_counts = df['Panic Attack'].value_counts()

# Create a list of labels for the pie chart
labels = ['Depression', 'Anxiety', 'Panic Attack']

# Create a list of counts for each category
counts = [depression_counts[1], anxiety_counts[1], panic_counts[1]]

# Set the color scheme for the pie chart
colors = ['#FF6F61', '#FFB07C', '#FFD700']

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(counts, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Distribution of Depression, Anxiety, and Panic Attack')
plt.show()
```

```python
# Filter the DataFrame for 'CSE Data Science' students
cse_ds_df = df[df['Course'] == 'CSE Data Science']

# Bar Chart - Year of Study
year_counts = cse_ds_df['Year of Study'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(year_counts.index, year_counts.values)
plt.title('Year of Study Distribution for CSE Data Science Students')
plt.xlabel('Year of Study')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram - CGPA
plt.figure(figsize=(8, 6))
plt.hist(cse_ds_df['avg_cgpa'], bins=10, edgecolor='black')
plt.title('Average CGPA Distribution for CSE Data Science Students')
plt.xlabel('Average CGPA')
plt.ylabel('Count')
plt.show()
```

### Inference:
